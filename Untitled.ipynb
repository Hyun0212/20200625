
## [문제] subplot 을 이용해서 2행 1열 구조 2개씩
- 4가지 색상 다르게.
- 4가지 스타일 다르게
- 4가지 마커 서로 다르게
- 범례 표시
- x축, y축, 제목

x=0부터 5까지 단위 1
y1= x+1
y2= x+2
y3= x+3
y4= x+4

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,5,1)
y1= x+1
y2= x+2
y3= x+3
y4= x+4
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,'r--',marker ='*')
plt.plot(x,y2,'g--',marker ='+')
plt.xlabel('x')
plt.ylabel('y')
plt.title('subplot')
plt.legend(['y1','y2'])
plt.subplot(1,2,2)
plt.plot(x,y3,'y--',marker ='o')
plt.plot(x,y4,'b--',marker ='v')
plt.xlabel('x')
plt.ylabel('y')
plt.title('subplot')
plt.legend(['y3','y4'])
```
